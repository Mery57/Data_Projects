## G. Connection au serveur PostgreSQL

### G.1. Connecter a la base de donnee M2i

```python
#importer librairies
import psycopg2
import psycopg2.extras as extras
```

```python
#connecter a la base de donnee M2i
try:
    conn = psycopg2.connect("dbname=M2I_Test user=postgres password=1234")
except (Exception, psycopg2.DatabaseError) as error:
    print(error)
    print("Probleme de connexion")
    
print("Succes de Connexion")
```

```python
##creer un curseur
cur = conn.cursor()
```

### G.2. Creer table airbnb_paris_listings_m2i

```python
# creer table airbnb_paris_listings_m2i avec l'id
requete = 'create table airbnb_paris_listings_m2i(id decimal primary key not null )'

#executer la requete
cur.execute(requete)
conn.commit()
```

```python
#dictionnaire de types
dict_type={'int64':'decimal', 'object':'text', 'float64':'real', 'datetime64[ns]':'date'}
dict_type['int64']
```

```python
# prepare requete pour ajouter de nouvelles colonnes a la table airbnb_paris_listings_m2i
table_name = 'airbnb_paris_listings_m2i'
requete = 'alter table ' + table_name 
for i,attribut in enumerate(df.columns):
    if(df[attribut].dtype)=="datetime64[ns]":
        print(attribut)
    if(i==0):
        continue
    requete += ' add column ' + attribut+ " " + dict_type[str(df[attribut].dtype) ] 
    if(i<len(df.columns)-1):
        requete+=','+ '\n'
```

```python
print(requete)
```

```python
#Executer la requete 
cur.execute(requete)
conn.commit()
```

### G.3. Inserer les donnees dans la table

```python
#sauvegrader les donnees du dataframe dans une liste de tuples
tuples = [tuple(x) for x in df.to_numpy()]
```

```python
#executer la requete convenable pour inserer les donnees dans la table
try:
    requete = "insert into " + table_name + " values %s" 
    extras.execute_values(cur, requete,  tuples)
    conn.commit()
except (Exception, psycopg2.DatabaseError) as error:
    print('Error {}'.format(error))
    conn.rollback()
    cur.close()
print("execute_values() done")
```

```python
#lire la prmemiere ligne de la table 
requete = 'Select * from '+ table_name+ ' limit 1'
cur.execute(requete)
conn.commit()
res = cur.fetchall()
for row in res:
    print(row)
```
